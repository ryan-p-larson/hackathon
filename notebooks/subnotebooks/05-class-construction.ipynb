```javascript
function Knowledge(courses) {
    this.courses = courses;
    
    this.getSubset = function(attr, value) {
        var filtered_courses = d3.values(this.courses).filter(d => {
            return d[attr] === value;
        })
        .map(d => d.courseID);
        return filtered_courses;
    };
}
```

```javascript
function KnowledgeGraph(knowledge) {
    // attrs
    this.knowledge = knowledge;
    this.nodes = [];
    this.links = [];
    
    
    this.addNode = function(nodeID) {
        this.nodes.push(nodeID);
    }
    
    this.addLink = function(link) {
        this.links.push(link);
    }
    
    this.isPermissible = function(nodeID) {
        return this.nodes.indexOf(nodeID) > -1;
    };
    
    this.dedupeNodes = function() {
        this.nodes = _.uniq(this.links, d => d.courseID);
        /*
        this.nodes.filter(function(value, index, self) { 
            return self.indexOf(value) === index;
        });
        */
    }
    
    this.dedupeLinks = function() {
        //_.uniq(array, element => element.name + element.age);
        this.links = _.uniq(this.links, d => d.source + d.target);
    }
    
    this.clearNodes = function() {
        this.nodes = [];
    }
    
    this.clearLinks = function() {
        this.links = [];
    }
    
    this.updateGraphLinks = function() {
        // save obj for later
        var kg = this;
        
        // Clear links
        this.clearLinks();
        
        // For each node, add it's permissible prereqs
        this.nodes.forEach(d => {
            var node = this.knowledge.courses[d];
            
            // FIX THIS
            // right now it checks node , shouldnt have to
            if (node !== undefined) {
              node['before']
                .filter(e => { return kg.isPermissible(e.source); })
                .forEach(e => { kg.addLink(e, d); });
            };
            //
            
        });
        
        // remove duplicate links
        this.dedupeLinks();
    }
    
    this.cloneGraphNodes = function() {
        return this.nodes.map(d => _.cloneDeep(this.knowledge.courses[d]));
    };
    
    this.createGraphFromList = function(node_list) {
        // obg for later
        var kg = this;
        
        // start fresh
        this.clearNodes();
        this.clearLinks();
        
        // add all nodes from list and then update links
        node_list.forEach(d => kg.addNode(d));
        this.dedupeNodes();
        
        this.updateGraphLinks();
        
        return {nodes: this.cloneGraphNodes(), links: this.links};
    };
    
    this.createGraphFromSearch = function(nodeID) {
        var kg = this;
        
        // start fresh
        this.clearNodes();
        this.clearLinks();
        
        // traversal queue
        var queue = [];
        queue.push(nodeID);
        
        // Search graph, adding nodes to our list along the way
        while (queue.length > 0) {
            var node = this.knowledge.courses[queue.pop()];
            
            if (node !== undefined) {
                // add node to our list
                kg.addNode(node.courseID);
                
                // add before
                node['before'].forEach(e => { 
                    // only push node if it's not permissible, AKA we haven't seen it before
                    if (kg.isPermissible(e['source']) === false) { queue.push(e['source']); }
                });
                
                // Add antecedents(after)
                // to-do
            };
        };
        
        // dedupe nodes before creating links
        this.dedupeNodes();
        this.updateGraphLinks();
        
        return {nodes: this.cloneGraphNodes(), links: this.links};
    }
    
}
```

```javascript
function draw_chart(graph) {
    
    // sanity check
    if (graph.links.length < 1 || graph.nodes.length < 1) {
        console.log("No links, skipping.");
        return;
    }
    
    var width = 960,
        height = 600,
        svg = svgDOM(width, height);
    
    var departments = graph.nodes
        .map(d => { return d.department; })
        .filter(function(value, index, self) { 
            return self.indexOf(value) === index;
        });
    var color_scale = d3.scaleOrdinal(d3.schemeCategory10)
        .domain(departments);

    var sankey = d3_sankey.sankey()
        .nodeId(function(d) { return d.courseID; })
        .nodeAlign(d3_sankey.sankeyCenter)
        .nodeWidth(20)
        .nodePadding(15)
        .size([width, height]);
    
    sankey(graph);

    var link = svg.append("g")
        .attr("class", "links")
        .attr("fill", "none")
        .attr("stroke", "#000")
        .attr("stroke-opacity", 0.2)
      .selectAll("path");
    var node = svg.append("g")
        .attr("class", "nodes")
        .attr("font-family", "sans-serif")
        .attr("font-size", 10)
      .selectAll("g");

    link = link
        .data(graph.links)
        .enter().append("path")
        .attr("d", d3_sankey.sankeyLinkHorizontal())
        .attr('class', 'link')
        .attr("stroke-width", function(d) { return Math.max(1, d.width); });
    link.append("title")
        .text(function(d) { return d.source.name + " → " + d.target.name; });

    node = node
        .data(graph.nodes)
        .enter().append("g")
        .filter(function(d) { return (d['targetLinks'].length > 0) || (d['sourceLinks'].length > 0); })
        .attr('class', 'node');
    node.append("rect")
        .attr('id', d => { return d.courseID; })
        .attr('class', 'node')
        .attr("x", function(d) { return d.x0; })
        .attr("y", function(d) { return d.y0; })
        .attr("height", function(d) { return d.y1 - d.y0; })
        .attr("width", function(d) { return d.x1 - d.x0; })
        .attr("fill", function(d) { return color_scale(d.department); })
        .attr("stroke", "#000");
    var course_names = node.append("text")
        .attr("x", function(d) { return d.x0 - 6; })
        .attr("y", function(d) { return (d.y1 + d.y0) / 2; })
        .attr("dy", "0.35em")
        .attr('class', 'courseName')
        .attr("text-anchor", "end")
        .text(function(d) { return d.name; })
      .filter(function(d) { return d.x0 < width / 2; })
        .attr("x", function(d) { return d.x1 + 6; })
        .attr("text-anchor", "start");
    node.append("title")
        .text(function(d) { return d.name + "\n" + d.courseID +"\n" + d.department; });

    //return svg;
    return $$.svg(svg.node().outerHTML);
}
```
