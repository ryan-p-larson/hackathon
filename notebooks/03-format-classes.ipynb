```python
def format_course(x):
    return ":".join(x.split(':')[:2])

def format_name(x):
    human = x.split(' ', 1)[1].strip()
    return human.replace('\n', '').strip()

def format_prereqs(prereq):
    if (prereq == None):
        return None
    elif (prereq.startswith('None')):
        return None
    
    words = prereq.split()
    courses = [word.replace('(', '').replace(')', '') for word in words if ':' in word]
    
    return courses

```

```python
def clean_df(df):
    df['Name'] = df['Name'].apply(lambda x: format_name(x))
    df['Course'] = df['Course'].apply(lambda x: format_course(x))
    df['Prerequisites'] = df['Prerequisites'].apply(lambda x: format_prereqs(x))
    
    return df[['Name', 'Course', 'Prerequisites']]
```

```python
def get_department_df(department):
    
    dep_scraper = DepartmentScraper()
    
    departmentID = departments[department]
    # get all course links
    dep_scraper.start_scrape(departmentID)
    # scrape all courses
    cour_scraper = CourseScraper(dep_scraper.class_pages)
    
    return clean_df(cour_scraper.courses_df)
```
